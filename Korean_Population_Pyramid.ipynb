# 1. Population Pyramid

## Male Population Defined

```python
import pandas as pd
df_m = pd.read_excel('201108_population_by_age.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951 (from string to integer)
df_m.head(3)
```

## Female Population Defined

```python
df_w = pd.read_excel('201108_population_by_age.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns # making the column names identical
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951 (from string to integer)
df_w.head(3)
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0]//1000) # Unit by 1,000
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.title('2011 Korean Population Pyramid')
plt.savefig('2011_KOpopulation.png', dpi=100)
plt.show()
```

```python
df_m_2 = pd.read_excel('202108_population_by_age.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_w_2 = pd.read_excel('202108_population_by_age.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_m_2.iloc[0] = df_m_2.iloc[0].str.replace(',', '').astype(int)
df_w_2.iloc[0] = df_w_2.iloc[0].str.replace(',', '').astype(int)
df_w_2.columns = df_m_2.columns
```

```python
df_m_2.head(3)
```
