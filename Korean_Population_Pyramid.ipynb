## Male Population Defined

```python
import pandas as pd
df_m = pd.read_excel('201108_population_by_age.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951 (from string to integer)
df_m.head(3)
```
